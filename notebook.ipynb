```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('Zara_sales_EDA.csv',delimiter=';')
```

```python
df = pd.DataFrame(data)
```

```python
df.drop('url',axis=1,inplace=True) #The column url is not important for us that's why we are removing it
```

```python
df.head(1)
```

```python
df.dropna(inplace=True) #Removing the missing values
```

```python
df.info()
```

```python
df[['Sales Volume','price']].describe()
```

```python
df.duplicated().sum() #checking for duplicate values
```

There is no duplicate value

```python
plt.bar(x=df['origin'],height=df['Sales Volume'],data=df)
plt.xticks(rotation=45)
plt.grid()
plt.show()
```

The country of origin not making any effects on sales valume they are almost equal

```python
plt.bar(x=df['Promotion'],height=df['Sales Volume'],data=df)
plt.xticks(rotation=45)
plt.grid()
plt.show()
```

This show that the product which are included in promotion they have more sales volume a positive correlation

```python
plt.bar(x=df['Product Position'],height=df['Sales Volume'],data=df)
plt.xticks(rotation=45)
plt.grid()
plt.xlabel('Product Position')
plt.ylabel('Sales Volume')
plt.show()
```

The product position not making any effects on sales valume they are almost equal

```python
plt.bar(x=df['Product Position'],height=df['price'],data=df)
plt.xticks(rotation=45)
plt.grid()
plt.xlabel('Product Position')
plt.ylabel('Price')
plt.show()
```

```python
plt.bar(x=df['material'],height=df['Sales Volume'],data=df)
plt.xticks(rotation=45)
plt.grid()
plt.xlabel('Material')
plt.ylabel('Sales Volume')
plt.show()
```

The material of cloth also not making any effect on sales they all are almost equal

```python
plt.bar(x=df['material'],height=df['price'],data=df)
plt.xticks(rotation=45)
plt.grid()
plt.xlabel('Material')
plt.ylabel('Price')
plt.show()
```

The Satin,Silk material have low price as compared to others material

```python
plt.bar(x=df['Seasonal'],height=df['Sales Volume'],data=df)
plt.xticks(rotation=45)
plt.grid()
plt.xlabel('Seasonal')
plt.ylabel('Sales Volume')
plt.show()
```

The product is seasonal or not there is no effect on sales

```python
plt.pie(x=df['origin'].value_counts(),labels=df['origin'].value_counts().index,autopct='%1.2f%%')
plt.show()
```

Maximum product were origin from China almost 20%

```python
plt.pie(x=df['Product Position'].value_counts(),labels=df['Product Position'].value_counts().index,autopct='%1.2f%%')
plt.show()
```

More product displayed on Aisle position followed by End-cap and Front of store

```python
plt.bar(x=df['section'],height=df['Sales Volume'],data=df)
plt.xticks(rotation=45)
plt.grid()
plt.xlabel('Section')
plt.ylabel('Sales Volume')
plt.show()
```

Woman Section has more sales as compared to man section
